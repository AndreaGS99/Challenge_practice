# Informe de análisis



Derivado de analisis de datos procesado comparativo de las cuatro tiendas tuvo como objetivo identificar cuál presenta el menor desempeño, con el fin de recomendar su venta. Para ello, se evaluaron indicadores como los ingresos totales, las categorías y productos más vendidos, la calificación promedio de los clientes y el costo de envío. Los resultados muestran que la Tienda 4 registra la menor facturación entre todas, lo que indica un menor volumen de ventas y menor rendimiento comercial. Aunque presenta el costo de envío más bajo y una calificación promedio aceptable, estos factores no compensan su bajo nivel de ingresos en comparación con las demás tiendas. Por su parte, las tiendas 1, 2 y 3 muestran mejores resultados en facturación y desempeño general. En consecuencia, considerando todos los indicadores analizados, se recomienda que el Sr. Juan venda la **Tienda 4**, ya que es la que presenta el menor rendimiento dentro del conjunto evaluado.
